plotting multiple charts in a grid

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Subplots banayein
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# First plot
axes[0, 0].plot(years, apples, 's-b')
axes[0, 0].plot(years, oranges, 'o-r')
axes[0, 0].set_xlabel('year')
axes[0, 0].set_ylabel('yield (tons per hectare)')
axes[0, 0].legend(['apples', 'oranges'])
axes[0, 0].set_title('crop yields in kanto')

# Second plot
axes[0, 1].set_title('Sepal length vs. sepal width')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=100, 
                ax=axes[0, 1])

# Third plot
axes[0, 2].set_title('Distribution of sepal width')

# **Yahaan indentation ka dhyan de**
axes[0, 2].hist([setosa_df.sepal_width, 
                 versicolor_df.sepal_width, 
                 virginica_df.sepal_width], 
                bins=np.arange(2, 5, 0.25),  
                stacked=True);
# Pass the axes into seaborn
axes[0, 2].legend(['Setosa', 'Versicolor', 'Virginica'])

# Pass the axes into seaborn
axes[1, 0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1, 0])

# Pass the axes into seaborn
axes[1, 1].set_title('Flights traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1, 1])

# Plot an image using the axes
axes[1, 2].set_title('Data Science Meme')
axes[1, 2].imshow(img)
axes[1, 2].grid(False)
axes[1, 2].set_xticks([])
axes[1, 2].set_yticks([])

plt.tight_layout(pad=2)
axes[0, 2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
                bins=np.arange(2, 5, 0.25));

```

```python
def split_multicolumn(col_series):
    result_df=col_series.to_frame()
    options=[]
    #Iterate over the column
    for idx,value in col_series[col_series.notnull()].items():
        #break each value into list of options
        for option in value.split(';'):
        #add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option]=False
    #mark the value in the option column as True
            result_df.at[idx,option]=True
    return result_df[options]
```

```python
dev_type_df=split_multicolumn(survey_df.DevType)
```

```python
dev_type_df
```

```python
dev_type_totals=dev_type_df.sum().sort_values(ascending=False)
dev_type_totals
```

Inferences and Conclusions

we've drawn many interesting inferences from the survey,here's a summary of few of them:

1.Based on the demographics of the survey respondents, we can infer that the survey is somewhat representative of the overall programming community,although it definitely has fewer responses from programmers in non-english-speaking countries and from women and non-binary genders.

2.The programming community is not as diverse as it can be,and although things are improving,we should take more efforts to support and encourage members of underrepresented communities-wheather it is in terms of age,country,race,gender or otherwise.

3.Most programming hold a college degree,although a fairly large percentage did not have computer science as thier major in college, so computer science degree isn't compulsory for learning to code or bulding a carrer in programming.

4.A significant percentage of programmers either work part time or as a freelancers,and this can be a great way to break into field,epecially when ypu're just getting started.

5.Javascript and HTML/CSS are the most used programming languages in 2020,closely followed by SQL and Python.

6.python is the language most people are interested in learning-since it is an easy-to-learn genral purpose programming language well suited for variety of domains.

7.Rust and TypeScript are the most 'loved' language in 2020,both of which have small but fast-growing communities.python language is a closed third,despite already being a weidly used language.

8.programmers around the word sems be working for around 40 hours per week on average with slight variations by country.

9.you can learn and start programming professionally at any age,and you're likely to have a long and fulfilling carrer if you also enjoy programming as a hobby.
